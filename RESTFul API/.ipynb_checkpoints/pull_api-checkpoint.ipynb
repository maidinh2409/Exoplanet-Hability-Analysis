```python
from callAPI import CallAPI
from gcloud import UploadGCloud
```

```python
api = CallAPI()
url = "https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+pl_name,st_age,pl_masse,pl_rade,pl_dens,pl_eqt,pl_insol,pl_orbsmax,pl_orbper,hostname,st_teff,st_rad,st_mass,st_met,st_lum,discoverymethod,disc_year,disc_facility,disc_locale+from+pscomppars&format=json"

request = api.getJSON(url)
api.saveLocal(request, "pscomppars.json", "data")

cloud = UploadGCloud()

blob = "pscomppars.json"
bucket_name = "example_bucket_1304"
cloud.upload_nolocal(bucket_name, request, blob)
```
